```python
# Now Importing the installed libraries
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
url="https://en.wikipedia.org/wiki/List_of_active_Indian_military_aircraft"
```

```python
 # Here we are sending request to get the permission from IMDB website. For this we use request library.
response=requests.get(url)
```

```python
response                # when we get output for our request as 200 it means now we can scrap the website.
```

```python
response.content    # this command prints the HTML code of that website.
```

```python
# As we can see the data which we got is not in readable format so to make it more understable we use beatifulsoup library.
soup=BeautifulSoup(response.content,'html.parser')
```

```python
soup
```

```python
# we want only table data for that we used find function to get the table from class containing word "Wikitable"
table = soup.find('table', {'class': 'wikitable'})
```

```python
table
```

```python
headers = []
for th in table.find_all('th'):
    headers.append(th.text.strip())
```

```python
desired_headers = ["Aircraft", "Origin"]

# Find the indices of the desired headers
desired_indices = [headers.index(header) for header in desired_headers]
```

```python
desired_headers
```

```python
rows = []
for tr in table.find_all('tr')[1:]:  # Skip the header row
    cells = tr.find_all('td')
    if len(cells) > 0:
        row = {}
        for i in desired_indices:
            row[headers[i]] = cells[i].text.strip()
        rows.append(row)
```

```python
rows
```

```python
data=[]
for i in rows[0::2]:
  data.append(i)
```

```python
data
```

```python
df = pd.DataFrame(data)       # created a dataframe using pandas library
```

```python
df.to_csv("Active_Indian_Military_Aircraft.csv", index=False)      # converted our data to csv file.
```
